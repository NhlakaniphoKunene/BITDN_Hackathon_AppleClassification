```python
# ___Cell no. 12___
import sys
sys.path.append("..")
from source.utils import split #  a pre-defined function to split the data into training and testing
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
```

```python
# ___Cell no. 13___

Y = Y.map({'S': 1, 'B': 0})
```

```python
# ___Cell no. 14___

selected_indexes_sk = []
selected_indexes_v = []


for x_roll in X_arrayRol_sk:
    Xtrain_sk, Xtest_sk, Ytrain_sk, Ytest_sk  = split( x_roll, Y) # splitting the data
    print("(Number of samples, number of features) = ", Xtrain_sk.shape)
    sfs = SequentialFeatureSelector(estimator=LogisticRegression(solver = 'newton-cg'), n_features_to_select=10, direction = 'forward', scoring = 'precision' ) # def SFS
    sfs.fit(Xtrain_sk.values, Ytrain_sk)
    selected_indexes_sk.append(sfs.support_) # storing the indices of the best n_features_to_select columns
    
for x_roll in X_arrayRol_v:
    Xtrain_v, Xtest_v, Ytrain_v, Ytest_v  = split( x_roll, Y) # splitting the data
    print("(Number of samples, number of features) = ", Xtrain_v.shape)
    sfs = SequentialFeatureSelector(estimator=LogisticRegression(solver = 'newton-cg'), n_features_to_select=10, direction = 'forward', scoring = 'precision' ) # def SFS
    sfs.fit(Xtrain_v.values, Ytrain_v)
    selected_indexes_v.append(sfs.support_) # storing the indices of the best n_features_to_select columns
```
